# Contributing

## Overview

The Turkish Delight Corpus (TDC) is an open educational and research resource designed to support reproducible computational analysis of Turkish makam music.

Researchers, educators, students, and developers are encouraged to contribute improvements, educational materials, computational methods, and documentation that enhance the accessibility and sustainability of the project.

## Ways to Contribute

Contributions to the Turkish Delight Corpus may include:

- improving existing notebooks,
- developing new computational analyses,
- contributing visualization techniques,
- integrating additional open datasets,
- correcting documentation,
- reporting issues,
- improving reproducibility,
- extending educational materials.

## Coding Guidelines

To maintain consistency throughout the project, contributors are encouraged to:

- write well-documented Python code,
- follow PEP 8 coding conventions,
- develop reproducible computational workflows,
- include explanatory Markdown cells,
- generate figures programmatically,
- clearly document data preprocessing steps,
- appropriately cite external datasets and publications.

## Reproducibility

Reproducibility is a fundamental principle of the Turkish Delight Corpus.

Contributors should ensure that notebooks:

- execute sequentially without manual intervention,
- use openly available datasets whenever possible,
- avoid hidden preprocessing steps,
- document software dependencies,
- generate all analyses programmatically.

## Citation

If you use the Turkish Delight Corpus in research or teaching, please cite the corresponding publication (when available) together with the original datasets and software resources employed throughout the analyses.

Proper citation supports reproducible research and acknowledges the valuable contributions of the original dataset developers.

## Acknowledgements

The Turkish Delight Corpus builds upon the outstanding datasets, software tools, and computational musicology research developed by the Music Technology Group (MTG), Universitat Pompeu Fabra (UPF), together with contributions from the broader Music Information Retrieval community.

Their commitment to open science has made many of the analyses presented throughout this book possible.
