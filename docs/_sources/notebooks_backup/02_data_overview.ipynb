```python
import pandas as pd
from pathlib import Path

DATA_RAW = Path("../data/raw").resolve()

csv_file = DATA_RAW / "Turkish maqam music pieces in time series format_3000 rows x 365 columns.csv"

df = pd.read_csv(csv_file)

print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum().sum()

print("Total Missing Values:", missing_values)

missing_percentage = (
    missing_values /
    (df.shape[0] * (df.shape[1]-1))
) * 100

print(f"Missing Percentage: {missing_percentage:.2f}%")
```

```python
df.head()
```

```python
print("Missing values:", df.isnull().sum().sum())
```

```python
X = df.iloc[:,1:]

desc_stats = X.describe()

desc_stats
```

```python
X = df.iloc[:,1:]

print("Global Mean :", X.stack().mean())
print("Global Std  :", X.stack().std())
print("Global Min  :", X.min().min())
print("Global Max  :", X.max().max())
```

```python
print("Global Median :", X.stack().median())
```

```python
summary = pd.DataFrame({
    "Metric": [
        "Number of Pieces",
        "Time Steps",
        "Missing Values",
        "Global Mean",
        "Global Median",
        "Global Std",
        "Global Min",
        "Global Max"
    ],
    "Value": [
        len(df),
        X.shape[1],
        df.isnull().sum().sum(),
        round(X.stack().mean(),2),
        round(X.stack().median(),2),
        round(X.stack().std(),2),
        X.min().min(),
        X.max().max()
    ]
})

summary
```
