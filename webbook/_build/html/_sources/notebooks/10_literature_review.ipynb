## 1. Symbolic Music Corpora

The availability of publicly accessible symbolic music corpora has been one of the key factors enabling computational research on Turkish makam music.

The introduction of the SymbTr corpus represented the first comprehensive machine-readable symbolic collection specifically designed for Music Information Retrieval (MIR). It established a standardized representation of Turkish makam music scores and enabled studies on symbolic music analysis, melody processing, similarity analysis, and computational musicology {cite}`karaosmanoglu2012symbtr`.

The most recent release, SymbTr v3.0, contains approximately 3,000 symbolic scores spanning 164 makams, 135 usuls, and more than 1.2 million notes, making it the largest openly available symbolic corpus for Turkish makam music {cite}`karaosmanoglu2025symbtrv3`.

## 2. Phrase Segmentation and Melodic Analysis

Phrase-level analysis has become one of the most active research topics in computational Turkish makam music.

The introduction of expert-annotated phrase datasets enabled researchers to investigate melodic structure beyond complete musical works. These resources support phrase segmentation, melodic similarity analysis, motif discovery, and symbolic music retrieval.

Recent studies have demonstrated that phrase-level representations improve the understanding of local musical structure and facilitate the development of more sophisticated computational models.

## 3. Audio–Score Alignment

Audio–score alignment aims to synchronize symbolic music scores with recorded performances.

Research in Turkish makam music has introduced alignment techniques capable of connecting symbolic notation and audio recordings. These approaches have enabled musicological analysis, performance comparison, music education, and multimodal Music Information Retrieval.

The availability of aligned symbolic and audio representations has also facilitated the development of computational methods that integrate multiple musical modalities.

## 6. Emerging Research Directions

Recent advances in Artificial Intelligence have opened new research opportunities for computational Turkish makam music.

Promising future directions include foundation models for symbolic music, transformer-based architectures, graph neural networks, explainable artificial intelligence, multimodal learning, and cross-cultural computational musicology.

The integration of symbolic music representations with audio recordings, metadata, and large-scale pretrained models is expected to significantly influence future research in Music Information Retrieval.

## Summary

Computational research on Turkish makam music has evolved from symbolic corpus construction toward increasingly sophisticated machine learning and artificial intelligence methodologies.

The continuous development of openly available datasets, together with reproducible computational workflows, has established a solid foundation for future interdisciplinary research spanning computational musicology, Music Information Retrieval, machine learning, and digital humanities.

The Turkish Delight Corpus complements these efforts by providing an integrated educational platform that demonstrates modern computational analysis techniques using reproducible Jupyter notebooks.

The SymbTr corpus established one of the first comprehensive symbolic resources for Turkish makam music {cite}`karaosmanoglu2012symbtr`.
