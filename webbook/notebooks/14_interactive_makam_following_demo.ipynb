# Interactive Makam Following Demo

## Overview

The **Makam Following** web application is an interactive educational and research resource developed by **Rafael Caro Repetto** at the **Music Technology Group (MTG), Universitat Pompeu Fabra (UPF)**.

The application demonstrates real-time synchronization between symbolic musical scores and audio recordings of Turkish makam music. By combining score visualization with synchronized audio playback, it enables users to explore the relationship between symbolic notation and musical performance in an intuitive and interactive manner.

Unlike the reproducible computational analyses presented throughout the Turkish Delight Corpus (TDC), this application focuses on the real-time interaction between symbolic notation and musical performance. Together, these complementary resources provide a broader perspective on computational approaches to Turkish makam music.

The official **Makam Following** demonstration is available at:

## Educational Value

The Makam Following demonstration supports several important topics in computational musicology and Music Information Retrieval (MIR), including:

- Real-time score–audio synchronization
- Symbolic music representation
- Interactive music visualization
- Performance-aware music analysis
- Computational approaches to Turkish makam music

These topics complement the statistical analyses, visualization techniques, dimensionality reduction, clustering methods, and symbolic music processing workflows presented throughout the Turkish Delight Corpus.

---

## Relationship to the Turkish Delight Corpus

Although the Turkish Delight Corpus focuses primarily on reproducible computational analysis of symbolic music datasets, the Makam Following application provides an interactive performance-oriented perspective.

Together, the two resources illustrate complementary aspects of computational musicology.

| Turkish Delight Corpus | Makam Following |
|-------------------------|-----------------|
| Symbolic dataset analysis | Interactive score–audio synchronization |
| Statistical analysis | Real-time performance visualization |
| Data visualization | Interactive musical exploration |
| PCA and clustering | Performance-aware analysis |
| Reproducible Python workflows | Browser-based educational application |
