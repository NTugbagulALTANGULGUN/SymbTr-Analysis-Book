## SymbTr v3.0

SymbTr v3.0 is the primary symbolic corpus used throughout this book.

The dataset contains approximately 3,000 symbolic scores representing a wide variety of Turkish makams, usuls, musical forms, and composers.

Throughout this tutorial, SymbTr serves as the foundation for exploratory data analysis, statistical analysis, visualization, dimensionality reduction, clustering, and comparative computational musicology studies.

## Turkish Makam Symbolic Phrase Dataset

The Turkish Makam Symbolic Phrase Dataset extends the SymbTr corpus by providing manually annotated phrase boundaries created by expert musicians.

Instead of complete musical works, the dataset represents phrase-level symbolic units, enabling research on phrase similarity, phrase segmentation, symbolic music retrieval, and computational analysis of melodic structures.

The dataset has become an important benchmark for phrase-based computational musicology.

## Turkish Makam Melodic Phrase Dataset

The Turkish Makam Melodic Phrase Dataset provides manually annotated melodic phrase boundaries extracted from symbolic music scores.

These annotations facilitate research on melodic segmentation, melodic pattern discovery, phrase classification, and machine learning approaches for symbolic music analysis.

Together with the Symbolic Phrase Dataset, it complements the SymbTr corpus by supporting phrase-level computational investigations.

## Relationship Between the Datasets

Although these datasets are closely related, they represent different structural levels of Turkish makam music.

While SymbTr provides complete symbolic scores, the phrase datasets focus on localized musical structures that enable more detailed computational analyses.

Together they establish a unified framework for computational musicology research.

## Why Turkish Delight Corpus?

Unlike conventional dataset repositories, the Turkish Delight Corpus combines openly available datasets with fully reproducible computational workflows.

By integrating exploratory data analysis, statistical methods, visualization, machine learning, clustering, and comparative analysis into a unified Jupyter Book, the project serves both as an educational resource and as a reproducible research platform.

The Turkish Delight Corpus therefore complements existing Turkish makam music datasets by demonstrating not only **what** data are available but also **how** they can be analyzed using modern computational techniques.
